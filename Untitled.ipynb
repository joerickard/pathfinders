```python
def check_symmetric(matrix, n):
    problemEntries = list()
    for i in range(n):
        for j in range(i,n):
            if matrix[i][j] != matrix[j][i]:
                problemEntries.append((i,j))
        
    if len(problemEntries):
        return problemEntries
    else: 
        return 'is Symmetric'
     
```

```python
#get traffic*dist weight with some alpha weight on the traffic values

def weightAdd(traffic, distance, n, alpha):
    Weights = np.zeros((n,n))
    for i in range(n):
        for j in range(n):
            Weights[i][j] = (alpha * traffic[i][j]) * ((1-alpha) * distance[i][j])
    return Weights
```

```python
#find cost and shortest path from i to j
def findMinPath(i,j,weights):
    cost, path = sp.shortest_path(weights, return_predecessors=True)
    minPath = [j]
    while path[i][j] != i:
        minPath = [path[i][j]] + minPath
        j = path[i][j]
    minPath = [i] + minPath
    
    return (cost[i][j], [x+1 for x in minPath]) #+1 to account for indexing in our paper notes starting at 1.
```

Visualization should be joseph minard-esque. We want path width to represent throughput, color to represent recommended path, and length to represent distance. (x,y, $\theta$, length, width, color) to show segments from $(x_0, y_0) \rightarrow (x_1, y_1)$

```python
def drawMap(start, end):
    img = plt.imread('CampusMap.png')
    scale = 0.007
    fig, ax = plt.subplots(1,1,figsize=(img.shape[0]*scale, img.shape[1]*scale))
    ax.imshow(img)

    Nodes = [(620,240),(620,530),(740,310),(730,570),(740,710),(960,450),(950,680),(820,780),(740,900),(880,955),(930,955),(1140,670),(1030,820),(1180,750),(1170,955),(860,1220),(1150,960),(980,1460),(870,1495),(1340,960),(1090,1210),(1020,1520),(1210,1340),(1300,1330),(1450,1120),(1430,1490)]

    pathNodes = findMinPath(start, end, weights)[1]
    pathEdges = list()
    for i in range(len(pathNodes)-1):
        pathEdges.append((pathNodes[i]-1, pathNodes[i+1]-1))

    for i in range(weights.shape[0]):
        for j in range(i,weights.shape[1]):
            if weights[i][j]:
                rootNode = Nodes[i]
                endNode = Nodes[j]
                length = np.sqrt((rootNode[0] - endNode[0])**2 + (rootNode[1] - endNode[1])**2)
                color='gold'

                if rootNode[1]-endNode[1]: #no divide by zero
                    angle = np.degrees(np.arctan(np.abs(rootNode[0]-endNode[0])/np.abs(rootNode[1]-endNode[1])))
                    if rootNode[0] > endNode[0]: #in either Q2 or Q3
                        if rootNode[1] > endNode[1]: #in Q2

                            angle = 180 + angle
                        else: #in Q3
                            angle = 360 - angle
                    else: #in either Q1 or Q4
                        if rootNode[1] > endNode[1]:
                            angle = 180 - angle
                        else:
                            angle = angle
                else:
                    angle = 90


                if (i,j) in pathEdges or (j,i) in pathEdges:
                    color='blue'

                rect = patches.Rectangle((rootNode[1],rootNode[0]), 
                                         length, 
                                         (100 - weights[i][j])/2,
                                         angle=angle,
                                         linewidth=2,
                                         edgecolor=color,
                                         facecolor=color,
                                         alpha=0.4)
                ax.add_patch(rect)
```
